```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Download stopwords only (safe)
nltk.download('stopwords')

# Load CSV
df = pd.read_csv('/content/sentimentdataset.csv')

# Confirm columns
print("Available columns:", df.columns.tolist())

# Use correct column
text_col = 'Text'

# Define a simple tokenizer (no punkt)
def basic_tokenize(text):
    return text.split()

# Text cleaning function
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # Remove links
    text = re.sub(r'@\w+|\#', '', text)  # Remove @mentions, hashtags
    text = re.sub(r'[^a-z\s]', '', text)  # Keep only letters and spaces
    tokens = basic_tokenize(text)
    filtered = [word for word in tokens if word not in stopwords.words('english')]
    return " ".join(filtered)

# Apply cleaning
df['clean_text'] = df[text_col].apply(clean_text)

# Preview result
print(df[['Text', 'clean_text']].head())
```

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# TextBlob analysis
df['polarity'] = df['clean_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['clean_text'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# VADER analysis
analyzer = SentimentIntensityAnalyzer()
df['vader_compound'] = df['clean_text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Classify sentiment
def get_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['vader_compound'].apply(get_sentiment)

# Preview results
print(df[['Text', 'clean_text', 'polarity', 'subjectivity', 'vader_compound', 'sentiment']].head())
```

```python
# Step 8: Save cleaned dataset to new CSV
df.to_csv("cleaned_sentiment_data.csv", index=False)
```
